```python
# Goal: All books from all pages with 2-star rating
```

```python
base_url = "https://books.toscrape.com/catalogue/page-{}.html"
```

```python
base_url.format(40)
```

```python
page_num = 6
base_url.format(page_num)
```

```python
res = requests.get(base_url.format(1))
soup = bs4.BeautifulSoup(res.text,"lxml")
```

```python
products =soup.select(".product_pod")
```

```python
example = products[0]
```

```python
example.select(".star-rating.Three")
```

```python
example.select("a")[1]["title"]
```

```python
two_star_ratings = []

#loop through pages 1-50
for n in range(1,51):

  #get 50 different urls to scrape from
  scrape_url = base_url.format(n)
  #get the text from all 50 urls
  res = requests.get(scrape_url)

  #get a soup for each scraped url
  soup = bs4.BeautifulSoup(res.text,"lxml")
  #get all books on each scraped url page
  books = soup.select(".product_pod")

  #for each book on each scraped url check whether it includes a 2-star rating,
  #then grab the title and append it to the empty list
  for book in books:
    if len(book.select(".star-rating.Two")) != 0:
      book_title = book.select("a")[1]["title"]
      two_star_ratings.append(book_title)

#print out each individual title in the list
for i in two_star_ratings:
  print(i)
```
